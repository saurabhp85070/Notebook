    using namespace std;

    struct Node
    {
        int data;
        struct Node *next;
    } *head; // first node in list

    void showlist()
    {
        Node *ptr=head;
        cout<<"Address of first node(head node): "<<ptr<<endl;
        cout<<"List: ";
        while(ptr!=NULL)
        {
            cout<<ptr->data<<" ";
            ptr=ptr->next;
        }
    }

    int main()
    {
        head=new Node(); // allocating first node
        cout<<"Address of head node: "<<head<<endl;
        Node *n1=new Node(); 
        Node *n2=new Node();
        Node *n3=new Node();
        Node *n4=new Node();// allocating last node

        // So, we have total 5 nodes that will store five values

        head->data=2; // storing data in head node
        head->next=n1; // storing address of next node

        n1->data=3;
        n1->next=n2;

        n2->data=4;
        n2->next=n3;

        n3->data=5;
        n3->next=n4;

        n4->data=6;
        n4->next=NULL; // setting last address as NULL

        showlist();
        return 0;
    }

    // inserting a node at nth position
    #include<bits/stdc++.h>
    using namespace std;

    struct Node
    {
        int data;
        struct Node *next;
    } *head; 

    void showList() 
    {
        Node *temp=head;

        cout<<"List: ";
        while(temp!=NULL)
        {
            cout<<temp->data<<" ";
            temp=temp->next;
        }
        cout<<endl;
    }

    void insert(int data, int pos)
    {
        Node *n1=new Node();
        Node *ptr;
        n1->data=data;
        n1->next=NULL;

        if(pos==1)
        {
            n1->next=head; // old head will be our next node of current node
            head=n1; // making current node as our new head now
            return;
        }
        ptr=head;
        for(int i=0;i<pos-2;i++) // go till (n-1)th position
        {
            ptr=ptr->next;
        }// after end of loop ptr will will be pointing to (n-1)th node

        /* Lets understand:

        Now ptr->next have address of (n-1)th node
        But now current node should contain this address
        So, we need to do:
            n1->next=ptr->next;

        Till now we linked nth node with current, and we need to link(n-1)th node to current node
        So, we need to do:
            ptr->next=n1;

        */

        n1->next=ptr->next; 
        ptr->next=n1;
    }

    int main()
    {
        head=NULL;
        insert(8,1); // 8
        insert(12,2); // 8 12
        insert(99,1); // 99 8 12
        insert(44,2); // 99 44 8 12
        insert(66,1); // 66 99 44 8 12

        showList();
    }
