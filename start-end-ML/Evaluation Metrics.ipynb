# Regression metrics


Regression metrics are used to evaluate the performance of machine learning models that are designed for regression tasks, where the goal is to predict a continuous numerical value. These metrics provide quantitative measures to assess how well a regression model is able to make predictions on a given dataset. 

### Mean Absolute Error (MAE):

- MAE is less sensitive to outliers(outliers in the dataset have less impact on the MAE) compared to other metrics like MSE.
- If interpretability is important, MAE might be preferred as it provides error values in the same units as the target variable.
- MAE is often preferred when the presence of outliers is significant in the dataset, as it provides a more balanced measure of model performance.

### Mean Squared Error (MSE):

- It provides a measure of the overall accuracy of the model's predictions.
- It penalizes larger errors more heavily than smaller errors due to the squaring operation.
- MSE is commonly `used in optimization algorithms` for training regression models, as it provides a differentiable loss function that can be minimized efficiently.
- `R² score provides a straightforward interpretation of the overall fit of the model to the data`, `MSE offers a more detailed evaluation of prediction accuracy by focusing on the magnitude of errors`. Therefore, while MSE may be less interpretable in terms of explaining overall fit, it provides a more granular assessment of prediction accuracy, which can be valuable for model refinement and optimization.

### Root Mean Squared Error (RMSE):

RMSE is the square root of the MSE, providing an interpretable scale that is in the same units as the target variable.
It provides a measure of the average magnitude of errors in the same units as the target variable.
- RMSE is sensitive to outliers due to the squaring operation in MSE.
-  If outliers are present in the data, RMSE might be a more appropriate metric to use as it penalizes larger errors more heavily.
- Use RMSE when you want a metric that penalizes larger errors more heavily than smaller errors.

### R² Score (Coefficient of Determination):

R² score measures the proportion of variance in the dependent variable that is predictable from the independent variables.
- It indicates the goodness of fit of the model to the data, with `values closer to 1(R² score value range from 0 to 1) indicating a better fit`.
- Use R² when you want to understand how well the independent variables explain the variability of the dependent variable.
- R² score is particularly useful when you want to `compare different models in terms of their explanatory power` or when you need to `communicate the overall performance of the model` to stakeholders in a simple, interpretable manner.

### Summary:
- For a model that best fits your dataset, you would ideally want to achieve `low values for MAE, MSE, and RMSE` i.e. approaching zero.
- However, achieving an `MAE of exactly zero` is often unrealistic and may indicate overfitting.
- You would aim for a` high R² score`, indicating that the model explains a significant proportion of the variance in the dependent variable.
- However, it's important to note that a `high R² score` does not necessarily mean that the model is free from issues such as overfitting or bias.

# Classification Metrics

Classification metrics are used to evaluate the performance of machine learning models designed for classification tasks, where the goal is to predict categorical labels or classes for given input data. These metrics provide insights into how well a model is classifying instances into different classes. 

### Accuracy:

- Accuracy measures the proportion of correctly classified instances out of the total instances.
- Accuracy is suitable for balanced datasets where the classes are represented equally. However, it can be misleading in the presence of class imbalance.
- Use accuracy `when all classes are of equal importance `and there is `no class imbalance in the dataset`.



### Precision(Positive Predictive Value):

- It focuses on the accuracy of positive predictions. 
- Precision is important when the cost of false positives is high (e.g., in medical diagnosis or fraud detection).
- Use precision `when the focus is on minimizing false positives`.



### Recall (Sensitivity or True Positive Rate):

- It focuses on capturing all positive instances 
- Recall is crucial when the cost of false negatives is high (e.g., in disease detection or anomaly detection).
- Use recall `when the focus is on minimizing false negatives`.



### F1 Score:

- F1 score is the harmonic mean of precision and recall, providing a `balanced measure of a model's performance`.
- It considers both false positives and false negatives.
- F1 score is particularly useful `when there is an imbalance between the classes` or `when precision and recall are equally important` or `when both false positives and false negatives need to be minimized.
- Use F1 score `when you want a balance between precision and recall`.



### ROC Curve and AUC (Area Under the Curve):

- ROC curve plots the true positive rate (TPR) against the false positive rate (FPR) at various threshold settings.
- AUC summarizes the performance of the model across different threshold settings by calculating the area under the ROC curve.
- Higher AUC indicates better discrimination between classes.
- ROC-AUC is useful for evaluating binary classifiers, providing a single scalar value that represents the model's ability to discriminate between positive and negative classes.
- Use ROC curve and AUC `when you want to evaluate the performance of a binary classifier across different threshold settings`.

### Confusion Matrix:

- A confusion matrix is a table that summarizes the performance of a classification model by comparing actual and predicted class labels.
- It consists of four values: true positives (TP), false positives (FP), true negatives (TN), and false negatives (FN).
- From the confusion matrix, various metrics like accuracy, precision, recall, and F1 score can be calculated.
- Use the confusion matrix to `get a detailed breakdown of the model's performance across different classes`.

### Specificity:

- It complements recall 
- Specificity is particularly relevant in scenarios where correctly identifying negative instances is important.
- Use specificity `when correctly identifying negative instances is critical`.

### Jaccard Index

- The Jaccard index measures the similarity between two sets by calculating the size of the intersection divided by the size of the union of the sets.
- In the context of classification, particularly for evaluating the performance of multi-class classification models, the Jaccard index is used to measure the similarity between the predicted set of classes and the true set of classes for each instance.
- The Jaccard index ranges from 0 to 1, with 1 indicating perfect overlap between the predicted and true sets of classes.
- It's particularly useful when dealing with multi-label classification problems where instances can belong to multiple classes simultaneously.
- Use the Jaccard index when you are dealing with multi-class classification or multi-label classification problems.

### Log Loss

- Log loss measures the performance of a classification model by quantifying the difference between predicted probabilities and the actual binary outcomes.
- It's commonly used in binary and multi-class classification tasks where the model outputs probabilities for each class.
- Log loss penalizes incorrect classifications more heavily when the model's confidence is high and less heavily when the model's confidence is low.
- Lower log loss values indicate better model performance, with perfect predictions resulting in a log loss of 0.
- Log loss is more sensitive to the certainty of predictions and can penalize overly confident but incorrect predictions more severely.
- Use log loss `when you are working with classification tasks where the model outputs probabilities for each class`.
- If your model outputs probabilities and you want to assess its calibration and certainty in its predictions, log loss can be a suitable metric.

### ROC-AUC

The ROC curve is a graph that shows the trade-off between True Positive Rate (Recall) and False Positive Rate at different threshold levels.

- True Positive Rate (TPR) = Recall = Sensitivity → Measures how many actual positives are correctly classified.
- False Positive Rate (FPR) → Measures how many actual negatives are incorrectly classified as positive.

The ROC curve plots TPR vs. FPR for different classification thresholds.

**What is AUC (Area Under the ROC Curve)?**
- AUC is a single number that represents the overall performance of the classifier.
- It measures the area under the ROC curve, giving a value between 0 and 1:
    - AUC = 1.0 → Perfect classifier
    - AUC = 0.5 → Random guessing (no better than flipping a coin)
    - AUC < 0.5 → Worse than random (model is misclassifying)



**When to Use ROC-AUC?**
- ✅ Best for Binary Classification – When the model predicts probabilities (e.g., Spam vs. Not Spam).
- ✅ Compare Different Models – A higher AUC means better performance.
- ✅ Handle Imbalanced Datasets – Unlike accuracy, AUC works well even if the classes are unbalanced.

**📌 Example (Spam Detection Model)**
    
Suppose we build a spam classifier that predicts whether an email is spam or not.

- If we increase the threshold (only very confident predictions are labeled spam), we reduce FP but may also reduce TP.
- If we decrease the threshold, we catch more spam (increase TP) but may also increase FP.
    
The ROC curve helps us choose the best threshold by balancing TPR and FPR.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc

# Step 1: Create a sample dataset
np.random.seed(42)
X = np.random.rand(1000, 2)  # 1000 samples, 2 features
y = np.random.randint(0, 2, 1000)  # Binary target (0 or 1)

# Step 2: Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Train a classifier (Logistic Regression)
model = LogisticRegression()
model.fit(X_train, y_train)

# Step 4: Get probability predictions
y_prob = model.predict_proba(X_test)[:, 1]  # Probability for class 1

# Step 5: Compute ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

# Step 6: Compute AUC
roc_auc = auc(fpr, tpr)

# Step 7: Plot ROC Curve
plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Random guess line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

**Output Interpretation**
- The blue curve is the ROC curve.
- The AUC score (shown in the legend) tells how well the classifier performs.
- A higher AUC (closer to 1.0) means a better classifier.
- The gray dashed line represents random guessing (AUC = 0.5).

# Clustering metrics

Clustering metrics are used to evaluate the performance of clustering algorithms, which aim to partition a dataset into groups or clusters based on the similarity of data points. These metrics provide insights into how well the clusters are formed and how distinct they are from each other. 

### Inertia (Within-Cluster Sum of Squares):

- Inertia measures the sum of squared distances of samples to their closest cluster center.
- Lower inertia values indicate tighter and more compact clusters, as data points within each cluster are closer to their cluster center.
- It's commonly used in k-means clustering, where the objective is to minimize the inertia by adjusting the cluster centroids.
- Use inertia to evaluate the compactness of clusters and to assess how well data points within each cluster are grouped around their centroid.
- It's particularly useful in k-means clustering to assess the convergence of the algorithm and to compare different clustering solutions with varying numbers of clusters.

### Silhouette Score:

- Silhouette score measures how similar an object is to its own cluster compared to other clusters.
- It ranges from -1 to 1, where a high silhouette score indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters.
- Higher silhouette scores indicate better-defined clusters with distinct separation between clusters.
- Silhouette score is useful for evaluating the overall quality of clustering and determining the optimal number of clusters.
- Use silhouette score to assess the overall quality of clustering and to determine the optimal number of clusters.
- It helps in identifying well-defined clusters with clear separation between clusters and can guide the selection of the number of clusters that maximizes the silhouette score.

### Davies-Bouldin Index:

- Davies-Bouldin index measures the average similarity between each cluster and its most similar cluster, relative to the average dissimilarity between clusters.
- Lower Davies-Bouldin index values indicate better clustering, with clusters that are well separated and distinct from each other.
- It provides a measure of cluster compactness and separation, with lower values indicating more cohesive and distinct clusters.
- Use Davies-Bouldin index to measure the compactness and separation of clusters.
- It's useful for comparing the quality of clustering solutions obtained from different algorithms or parameter settings and for selecting the solution with the lowest Davies-Bouldin index.

### Calinski-Harabasz Index:

- Calinski-Harabasz index measures the ratio of between-cluster dispersion to within-cluster dispersion.
- Higher Calinski-Harabasz index values indicate better clustering, with more compact and well-separated clusters.
- It's based on the concept that good clusters have high inter-cluster similarity and low intra-cluster similarity.
- Use Calinski-Harabasz index to assess the overall compactness and separation of clusters.
- It helps in comparing the clustering solutions obtained from different algorithms or parameter settings and in selecting the solution with the highest Calinski-Harabasz index.

### Adjusted Rand Index (ARI) and Normalized Mutual Information (NMI):

- ARI and NMI are used to measure the similarity between two clusterings, taking into account both the similarities and differences in the assignments of data points to clusters.
- They provide a measure of clustering quality by comparing the clustering results to a ground truth or reference clustering, if available.
- ARI and NMI values range from 0 to 1, where higher values indicate better agreement between the clusterings.
- Use ARI and NMI to evaluate the similarity between the clustering results and a ground truth or reference clustering, if available.
- They help in assessing the accuracy and agreement of clustering solutions with known or expected cluster assignments, such as in evaluating clustering algorithms in supervised or semi-supervised settings.
