## What is Matplotlib?

Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

Matplotlib was created by John D. Hunter.

Matplotlib is open source and we can use it freely.

Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

The source code for Matplotlib is located at this github repository https://github.com/matplotlib/matplotlib

```python
import matplotlib

print(matplotlib.__version__)
```

Matplotlib graphs your data on Figures (i.e., windows, Jupyter widgets, etc.), each of which can contain one or more Axes (i.e., an area where points can be specified in terms of x-y coordinates (or theta-r in a polar plot, or x-y-z in a 3D plot, etc.). The most simple way of creating a figure with an axes is using `pyplot.subplots`. We can then use `Axes.plot` to draw some data on the axes:

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots() # Create a figure containing a single axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]) # Plot some data on the axes.
```

In fact, you can do the same in Matplotlib: for each Axes graphing method, there is a corresponding function in the `matplotlib.pyplot` module that performs that plot on the "current" axes, creating that axes (and its parent figure) if they don't exist yet. So the previous example can be written more shortly as:

```python
 plt.plot([1, 2, 3, 4], [1, 4, 2, 3]) # Matplotlib plot
```

 A figure can contain any number of Axes, but will typically have at least one. The easiest way to create a new figure is with pyplot:

```python
fig = plt.figure() # an empty figure with no Axes
fig, ax = plt.subplots() # a figure with a single Axes
fig, axs = plt.subplots(2, 2) # a figure with a 2x2 grid of Axes
```

It's convenient to create the axes together with the figure, but you can also add axes later on, allowing for more complex axes layouts.

###  The object-oriented interface and the pyplot interface

As noted above, there are essentially two ways to use Matplotlib:
- Explicitly create figures and axes, and call methods on them (the "object-oriented (OO) style").
- Rely on pyplot to automatically create and manage the figures and axes, and use pyplot functions for plotting

```python
# OO-style

x = np.linspace(0, 2, 100)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.

fig, ax = plt.subplots() # Create a figure and an axes.
ax.plot(x, x, label='linear') # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic') # Plot more data on the axes...
ax.plot(x, x**3, label='cubic') # ... and some more.
ax.set_xlabel('x label') # Add an x-label to the axes.
ax.set_ylabel('y label') # Add a y-label to the axes.
ax.set_title("Simple Plot") # Add a title to the axes.
ax.legend() # Add a legend
```

```python
# pyplot-style

x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear') # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic') # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

Typically one finds oneself making the same plots over and over again, but with different datasets, which leads to needing to write specialized functions to do th eplotting. The recommended function signature is something like:

```python
def my_plotter(ax ,data1, data2, param_dict):
    """
    A helper function to make a graph
    
    Parameters
    ---------
    ax : Axes
        The axes to draw to
        
    data1 : array
            The x data
            
    data2 : array
            The y data
            
    param_dict : dict
                Dictionary of kwargs to pass to ax.plot
    
    Returns
    ------
    out : list
        list of artists added (Basically everything you can see on the figure is an artist (even the Figure, Axes, and Axis objects).)
    """
    
    out = ax.plot(data1,data2, **param_dict)
    return out
```

which you would then use as:

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, ax = plt.subplots(1, 1)
my_plotter(ax, data1, data2, {'marker': 'x'})
```

or if you wanted to have 2 sub-plots:

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```

## Plotting x and y points

The `plot()` function is used to draw points (markers) in a diagram.

By default, the `plot()` function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the **x-axis**.

Parameter 2 is an array containing the points on the **y-axis**.

If we need to plot a line from (1, 3) to (8, 10), we have to pass two arrays [1, 8] and [3, 10] to the plot function.

```python
# Draw a line in a diagram from position (1, 3) to position (8, 10):

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

**To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.**

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

**You can plot as many points as you like, just make sure you have the same number of points in both axis.**

```python
# Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10)

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

**If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 (etc., depending on the length of the y-points.**

So, if we take the same example as above, and leave out the x-points, the diagram will look like this:

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints) # The x-points in this example are [0, 1, 2, 3, 4, 5]
plt.show()
```

# Bar plot

Bar graphs are one of the most commonly used plots to display categorical data using rectangular bars. In Matplotlib, you can create bar plots using plt.bar().

```python
plt.bar(x, height, width, color, label)
```

- x: The categories (e.g., names of items).
- height: The height of each bar (e.g., values for each category).
- width (optional): The width of the bars. Default is 0.8.
- color (optional): The color of the bars.
- label (optional): The label for the bars, used for legends.

**Commonly Used Parameters:**
    
- x: Coordinates for the x-axis, usually categorical values.
- height: Heights (values) corresponding to each category in x.
- width: Width of each bar, default is 0.8.
- color: Colors for bars, can be a single color or a list of colors.
- label: Label for the bars, useful for adding legends.
- align: Alignment of bars ('center' or 'edge'), default is 'center'.
- edgecolor: Color of the bar borders.
- linewidth: Width of the bar borders.
- alpha: Transparency level of the bars (0 to 1).

```python
import matplotlib.pyplot as plt

x=['python', 'java', 'C', 'C++']
y=[85,70,60,82]

plt.xlabel("language")
plt.ylabel("number")
plt.title("languaage vs number")

plt.bar(x, y)
plt.show()
```

## Different bar plots

### Side-by-Side Bars

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for x, y, and z
x = np.array([1, 2, 3, 4])  # X-axis values (categories or labels)
y = np.array([10, 20, 15, 25])  # Y-axis values corresponding to X
z = np.array([5, 18, 12, 22])  # Z-axis values corresponding to X

# Set the width of the bars
bar_width = 0.3

# Set the position of the bars on the X axis
# r1 is the position for the 'y' bars
# r2 is the position for the 'z' bars, offset by the width of one bar
r1 = np.arange(len(x))  # Positions for the first set of bars
r2 = [r + bar_width for r in r1]  # Offset positions for the second set of bars

# Create bar plot for 'y' values (first set of bars)
plt.bar(r1, y, color='blue', width=bar_width, edgecolor='grey', label='Y data')

# Create bar plot for 'z' values (second set of bars, side-by-side with the first set)
plt.bar(r2, z, color='green', width=bar_width, edgecolor='grey', label='Z data')

# Add labels for the X-axis
plt.xlabel('X values', fontweight='bold')

# Add custom tick marks for the X-axis, aligning with the center of the groups of bars
plt.xticks([r + bar_width / 2 for r in range(len(x))], x)

# Add Y-axis label
plt.ylabel('Values')

# Add legend to differentiate between the Y and Z bars
plt.legend()

# Show the plot
plt.show()
```

The bars for `y` and `z` are next to each other, which makes it easy to compare their values for each `x`.

### Stacked Bars

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for x, y, and z
x = np.array([1, 2, 3, 4])  # X-axis values (categories or labels)
y = np.array([10, 20, 15, 25])  # Y-axis values corresponding to X
z = np.array([5, 18, 12, 22])  # Z-axis values corresponding to X

# Create bar plot for 'y' values
plt.bar(x, y, color='blue', edgecolor='grey', label='Y data')

# Create bar plot for 'z' values stacked on top of 'y'
plt.bar(x, z, bottom=y, color='green', edgecolor='grey', label='Z data')

# Add labels for the X-axis
plt.xlabel('X values', fontweight='bold')

# Add Y-axis label
plt.ylabel('Values')

# Add legend to differentiate between the Y and Z bars
plt.legend()

# Show the plot
plt.show()
```

The bars for `z` are stacked on top of `y`, so the total height of the stacked bars represents the sum of `y` and `z` for each `x`.

### Horizontal Bar Plot

Instead of vertical bars, you can create horizontal bars to display your data. This can be useful if your category labels are long and you want to improve readability.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 15, 25])

# Create horizontal bar plot
plt.barh(x, y, color='purple', edgecolor='grey')

# Add labels
plt.xlabel('Values')
plt.ylabel('X values')

# Show plot
plt.show()
```

### Grouped Bar Plot

Similar to side-by-side bar plots but more specifically for multiple categories within each `x` value.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 15, 25])
z = np.array([5, 18, 12, 22])

# Bar width and positions
bar_width = 0.3
r1 = np.arange(len(x))
r2 = [r + bar_width for r in r1]

# Create grouped bar plot
plt.bar(r1, y, width=bar_width, color='blue', label='Y data')
plt.bar(r2, z, width=bar_width, color='orange', label='Z data')

# Add labels and ticks
plt.xlabel('X values', fontweight='bold')
plt.xticks([r + bar_width / 2 for r in range(len(x))], x)

# Add legend
plt.legend()

# Show plot
plt.show()
```

### Stacked Horizontal Bar Plot

A horizontal version of the stacked bar plot, which may work better for certain types of categorical data.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 15, 25])
z = np.array([5, 18, 12, 22])

# Create stacked horizontal bar plot
plt.barh(x, y, color='blue', label='Y data')
plt.barh(x, z, left=y, color='green', label='Z data')

# Add labels
plt.ylabel('X values', fontweight='bold')
plt.xlabel('Values')

# Add legend
plt.legend()

# Show plot
plt.show()
```

### 100% Stacked Bar Plot

In this plot, each stacked bar adds up to 100%, which is useful for comparing proportions between categories.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 15, 25])
z = np.array([5, 18, 12, 22])

# Normalize the data to percentages
total = y + z
y_percentage = y / total * 100
z_percentage = z / total * 100

# Create 100% stacked bar plot
plt.bar(x, y_percentage, color='blue', label='Y data')
plt.bar(x, z_percentage, bottom=y_percentage, color='green', label='Z data')

# Add labels
plt.xlabel('X values', fontweight='bold')
plt.ylabel('Percentage')

# Add legend
plt.legend()

# Show plot
plt.show()
```

### Segmented Bar Plot (Multiple Variables)

You can plot bars that are segmented based on different subcategories, useful in visualizing more detailed breakdowns.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.array([1, 2, 3, 4])
y1 = np.array([10, 20, 15, 25])
y2 = np.array([5, 10, 8, 12])
y3 = np.array([3, 7, 5, 9])

# Create segmented bar plot
plt.bar(x, y1, color='blue', label='Y1 data')
plt.bar(x, y2, bottom=y1, color='orange', label='Y2 data')
plt.bar(x, y3, bottom=y1 + y2, color='green', label='Y3 data')

# Add labels
plt.xlabel('X values')
plt.ylabel('Values')

# Add legend
plt.legend()

# Show plot
plt.show()
```

### Bar Plot with Error Bars

### Bar Plot with Multiple Categories (Using Pandas and Seaborn)

# Scatter plot

A scatter plot is used to visualize the relationship between two continuous variables. It consists of points plotted on a graph where the x and y coordinates represent values from two datasets. Scatter plots are helpful in identifying correlations, trends, and patterns.

```python
plt.scatter(x, y, s, c, marker, alpha, label)
```

**Commonly Used Parameters in `plt.scatter()`:**
    
- x: The data for the x-axis.
- y: The data for the y-axis.
- s (optional): The size of the markers (default is 20).
- c (optional): The color of the markers. This can be a single color or an array that maps to colormap values.
- marker (optional): Shape of the markers (e.g., 'o' for circles, '^' for triangles, etc.).
- alpha (optional): Transparency of the markers, where 0 is fully transparent, and 1 is fully opaque.
- label (optional): Label for the plot, used for legends.
- cmap (optional): Colormap for coloring based on a third variable.
- edgecolors (optional): The color of the marker borders.

# fill_between()

The `fill_between()` function in Matplotlib is used to fill the area between two horizontal curves (or between a curve and a baseline) on a plot. This is particularly useful for highlighting areas under a curve or between two curves, making it easier to visualize data trends.

```python
plt.fill_between(x, y1, y2=0, where=None, interpolate=False, color=None, alpha=0.5)
```

**Common Parameters:**

- x: The x-coordinates of the data points.
- y1: The y-coordinates of the first curve (top curve).
- y2 (optional): The y-coordinates of the second curve (bottom curve). If not specified, it defaults to 0 (the x-axis).
- where (optional): A boolean array that specifies where to fill the area.
- interpolate (optional): If True, it interpolates the fill between y1 and y2 to fill regions where they may not intersect with the x-axis.
- color (optional): Color of the fill.
- alpha (optional): Transparency level of the fill (0 is transparent, 1 is opaque).

```python
# Fill Between Two Curves

import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creating the plot
plt.plot(x, y1, label='sin(x)', color='blue')
plt.plot(x, y2, label='cos(x)', color='orange')

# Filling the area between the curves
plt.fill_between(x, y1, y2, where=(y1 > y2), color='lightblue', alpha=0.5, label='Area between sin and cos')

# Adding labels and title
plt.title("Fill Between Example")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Fill Between a Curve and the X-axis

import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creating the plot
plt.plot(x, y, label='sin(x)', color='blue')

# Filling the area between the curve and the x-axis
plt.fill_between(x, y, color='lightblue', alpha=0.5)

# Adding labels and title
plt.title("Fill Between Curve and X-axis")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', lw=0.5)  # Adding x-axis
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Fill Between Two Curves with Conditions

import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = 0.5 * np.sin(2 * x)

# Creating the plot
plt.plot(x, y1, label='sin(x)', color='blue')
plt.plot(x, y2, label='0.5 * sin(2x)', color='orange')

# Filling the area where y1 > y2
plt.fill_between(x, y1, y2, where=(y1 > y2), color='lightblue', alpha=0.5)

# Adding labels and title
plt.title("Conditional Fill Between Two Curves")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

# Stair or Step plot

A stair plot (or step plot) is a type of plot where data points are connected with horizontal and vertical lines, creating a step-like appearance. It is particularly useful for representing data that changes at discrete intervals, such as categorical data, digital signals, or histograms.

In Matplotlib, a stair plot can be created using the `plt.step()` function.

```python
plt.step(x, y, where='pre', label=None, color=None, alpha=None)
```

**Common Parameters:**

- x: The x-coordinates of the data points.
- y: The y-coordinates of the data points.
- where: Specifies where the steps should occur:
    - 'pre': Step change occurs before the x-value.
    - 'mid': Step change occurs at the x-value (centered).
    - 'post': Step change occurs after the x-value.
- label (optional): Label for the plot, used in the legend.
- color (optional): Color of the step lines.
- alpha (optional): Transparency level of the lines (0 is transparent, 1 is opaque).

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(1, 10)
y = np.random.randint(1, 10, size=9)

# Creating the stair plot
plt.step(x, y, where='pre', color='blue', label='Step plot')

# Adding labels, title, and legend
plt.title("Basic Step Plot (where='pre')")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(1, 10)
y = np.random.randint(1, 10, size=9)

# Creating the step plot with mid-alignment
plt.step(x, y, where='mid', color='orange', label='Step (mid)')

# Adding labels and title
plt.title("Step Plot (where='mid')")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(1, 10)
y = np.random.randint(1, 10, size=9)

# Creating the step plot with post-alignment
plt.step(x, y, where='post', color='green', label='Step (post)')

# Adding labels and title
plt.title("Step Plot (where='post')")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(1, 10)
y = np.random.randint(1, 10, size=9)

# Creating the step plot with custom styles
plt.step(x, y, where='pre', color='red', linestyle='--', alpha=0.8, label='Dashed Step')

# Adding labels, title, and legend
plt.title("Customized Step Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

**Use Cases of Stair Plot:**
    
- Cumulative Data: Often used in cumulative distribution functions or histograms to show a stepwise increase.
- Digital Signals: Perfect for visualizing changes in digital signals, as the steps mimic the behavior of on/off switches.
- Event-Driven Data: Useful when plotting data that changes at specific events or time intervals.
- Inventory Management: Displaying stock levels that change in steps based on orders or supplies.

**Key Differences in where Parameter:**
    
- where='pre': The step changes before each x-value (i.e., horizontal segment starts at the current x-value).
- where='mid': The step change occurs halfway between the current and the next x-value.
- where='post': The step changes after each x-value (i.e., horizontal segment starts after the current x-value).

# stairs()

The `stairs()` function in Matplotlib is another way to create step-like plots, but it provides more flexibility than `step()` in terms of customizing the appearance of the plot. It is primarily used to generate step graphs, where the data is represented as horizontal segments with vertical jumps between the steps.

```python
plt.stairs(values, edges, orientation='vertical', baseline=None, label=None, color=None)
```

**Common Parameters:**

- values: The y-values (heights) of the steps.
- edges: The x-values (edges) that define the positions of the steps.
- orientation (optional): Either 'vertical' (default) or 'horizontal', specifying the direction of the steps.
- baseline (optional): Sets the baseline of the stairs. Can be set to a value (like 0) or None.
- label (optional): Label for the plot, useful when displaying a legend.
- color (optional): Specifies the color of the steps.

```python
import matplotlib.pyplot as plt

# Sample data
values = [1, 2, 3, 4, 5]
edges = [0, 1, 2, 3, 4, 5]

# Creating the stairs plot
plt.stairs(values, edges, color='blue', label='Stairs plot')

# Adding labels, title, and legend
plt.title("Basic Stairs Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
values = [1, 2, 3, 4, 5]
edges = [0, 1, 2, 3, 4, 5]

# Creating the horizontal stairs plot
plt.stairs(values, edges, orientation='horizontal', color='green', label='Horizontal Stairs')

# Adding labels and title
plt.title("Horizontal Stairs Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
values = [3, 2, 4, 6, 1]
edges = [0, 1, 2, 3, 4, 5]

# Creating the stairs plot with baseline at y = 2
plt.stairs(values, edges, baseline=2, color='orange', label='Stairs with Baseline')

# Adding labels and title
plt.title("Stairs Plot with Baseline")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
values1 = [1, 2, 3, 4, 5]
edges1 = [0, 1, 2, 3, 4, 5]

values2 = [2, 3, 4, 5, 6]
edges2 = [0, 1, 2, 3, 4, 5]

# Creating two stair plots
plt.stairs(values1, edges1, color='red', label='Stairs 1')
plt.stairs(values2, edges2, color='blue', label='Stairs 2')

# Adding labels, title, and legend
plt.title("Multiple Stairs Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

**Key Differences Between `stairs()` and `step()`**

- stairs(): More flexible in defining the edges and baselines, providing additional control over how the step plot is drawn. It's typically more versatile for defining step edges and is preferred when you want specific control over the step placement and baseline.
- step(): Simpler and more commonly used for straightforward step plots where you want to define a single set of x and y values. It provides the where option ('pre', 'mid', 'post').

# Box plot

A box plot (also known as a box-and-whisker plot) is a graphical representation used to display the distribution of a dataset. It shows five main descriptive statistics: the minimum, the first quartile (Q1), the median (Q2), the third quartile (Q3), and the maximum. Box plots are especially useful for identifying outliers and comparing distributions across multiple datasets.

In Matplotlib, you can create box plots using the boxplot() function.

```python
plt.boxplot(data, notch=False, vert=True, patch_artist=False, labels=None, showmeans=False)
```

- data: The dataset(s) to plot. This can be a list of arrays or lists, one for each box.
- notch (optional): If True, creates a notched box plot to represent the confidence interval around the median.

- patch_artist (optional): If True, boxes are filled with color.
- tick_labels (optional): Labels for each dataset.
- showmeans (optional): If True, displays the mean in the box plot.

**Components of a Box Plot:**

- The Box: The box itself represents the interquartile range (IQR) from Q1 (25th percentile) to Q3 (75th percentile).
- Median: A line inside the box marks the median (Q2, or the 50th percentile).
- Whiskers: Lines extending from the box to the smallest and largest values, excluding outliers.
- Outliers: Data points outside 1.5 times the IQR from the quartiles, displayed as individual points.

# Violin plot

A violin plot is a method of plotting numeric data and is similar to a box plot, but it also shows the probability density of the data at different values. This allows you to see both the distribution of the data and summary statistics like the median, interquartile range, and outliers. The violin plot includes a kernel density estimation of the data, which gives it a distinct "violin-like" shape.

In Matplotlib, violin plots are created using the violinplot() function.

```python
plt.violinplot(dataset, vert=True, showmeans=False, showmedians=False, showextrema=True)
```

**Common Parameters:**

- dataset: The data to be plotted. It can be a list, array, or a list of arrays (for multiple violins).
- vert (optional): If True, plots the violins vertically (default). Set to False for horizontal violins.
- showmeans (optional): If True, displays the mean of the data.
- showmedians (optional): If True, displays the median of the data.
- showextrema (optional): If True, shows the minima and maxima of the data.
- bw_method (optional): The bandwidth of the kernel density estimation. You can adjust it to make the violin plot smoother or sharper.

**Components of a Violin Plot:**

- The "violin" shape: Represents the kernel density estimate of the distribution of the data.
- The central line: Shows the median or mean, depending on the plot's options.
- The thicker part: Represents areas where the data is more concentrated (higher probability density).
- The thinner part: Represents areas where the data is sparse.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random sample data
data = np.random.normal(0, 1, 100)

# Creating the violin plot
plt.violinplot(data)

# Adding title and labels
plt.title("Basic Violin Plot")
plt.ylabel("Value")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating sample data
data = [np.random.normal(0, 1, 100), np.random.normal(1, 1.5, 100)]

# Creating the violin plot with customizations
parts = plt.violinplot(data, showmeans=True, showmedians=True)

# Customizing the appearance of the violins
for pc in parts['bodies']:
    pc.set_facecolor('cyan')  # Fill color
    pc.set_edgecolor('black') # Border color
    pc.set_alpha(0.7)         # Transparency

# Adding title and labels
plt.title("Customized Violin Plot")
plt.ylabel("Value")
plt.xticks([1, 2], ['Data 1', 'Data 2'])

# Show the plot
plt.show()
```

**Key Points to Understand About Violin Plots:**

- Density Representation: The violin plot's shape represents the density of the data points. The width of the violin is proportional to the density, so wider areas indicate more data points at those values.
- Comparison with Box Plot: While box plots summarize data with just quartiles and outliers, violin plots provide a richer understanding of the distribution by showing the data's full density.
- Customizations: You can add statistical information like the mean, median, and extrema, as well as customize the appearance of the violins with colors, transparency, and line styles.
- Multiple Datasets: Violin plots are especially useful for comparing the distribution of multiple datasets side by side.

# Pie Chart

A pie chart is a circular chart divided into sectors, each representing a proportion of the whole. It's commonly used to show the relative sizes of data segments in a dataset. Each slice of the pie represents a category, and the size of the slice corresponds to the proportion of that category.

In Matplotlib, the pie() function is used to create pie charts.

```python
plt.pie(data, labels=None, colors=None, autopct=None, startangle=None, explode=None, shadow=False, counterclock=True)
```

**Common Parameters:**

- data: A list or array of numerical values representing the proportion of each slice.
- labels (optional): Labels for each slice.
- colors (optional): A list of colors to use for the slices.
- autopct (optional): Displays the percentage value of each slice. You can format it (e.g., '%1.1f%%' to show percentages with one decimal place).
- startangle (optional): The angle by which to start the pie chart (default is 0). You can rotate the start of the pie by setting an angle.
- explode (optional): A list or array that determines how far each slice is exploded (i.e., pulled out from the center of the pie). Values between 0 and 1.
- shadow (optional): If True, adds a shadow to the pie chart.
- counterclock (optional): If True, the slices are drawn counterclockwise; otherwise, they are drawn clockwise.

```python
import matplotlib.pyplot as plt

# Data to plot
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [30, 20, 40, 10]

# Creating the pie chart
plt.pie(sizes, labels=labels)

# Adding title
plt.title("Basic Pie Chart")

# Show the plot
plt.show()
```

**Key Points to Understand About Pie Charts:**

- Exploding Slices: You can explode slices to highlight a particular category.
- Limited Categories: Pie charts are best suited for datasets with relatively few categories. When there are too many categories, a pie chart can become cluttered and hard to interpret.

- Not for Exact Comparisons: Pie charts are not ideal for comparing exact values between categories. For more precise comparisons, bar charts are usually more effective.

# imshow()

The imshow() function in Matplotlib is used to display data as an image, i.e., on a 2D regular raster. It is commonly used for displaying images or visualizing 2D data (like heatmaps, matrices, or grayscale/color images).

# subplot()

The subplot() function in Matplotlib allows you to create multiple plots (or subplots) within a single figure. Its used to position multiple plots in a grid-like layout. This is particularly useful when you want to display several plots side-by-side or in rows/columns for easy comparison.

```python
plt.subplot(nrows, ncols, index)
```

- nrows: The number of rows in the grid.
- ncols: The number of columns in the grid.
- index: The position of the current plot in the grid (1-based index, counting left to right and top to bottom).

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plots
x = np.linspace(0, 10, 100)

# Creating a 2x2 grid of subplots
plt.figure(figsize=(8, 6))

plt.subplot(2, 2, 1)  # First plot in the grid (upper left)
plt.plot(x, np.sin(x))
plt.title("Sine Wave")

plt.subplot(2, 2, 2)  # Second plot (upper right)
plt.plot(x, np.cos(x), 'r')
plt.title("Cosine Wave")

plt.subplot(2, 2, 3)  # Third plot (lower left)
plt.plot(x, np.tan(x), 'g')
plt.title("Tangent Wave")

plt.subplot(2, 2, 4)  # Fourth plot (lower right)
plt.plot(x, -np.sin(x), 'b')
plt.title("Negative Sine Wave")

plt.tight_layout()  # Automatically adjust the subplots to fit in the figure area
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.gridspec import GridSpec

# Data for plots
x = np.linspace(0, 10, 100)

# Create a grid with customized row and column sizes
plt.figure(figsize=(10, 6))
gs = GridSpec(3, 3)  # 3 rows and 3 columns

# First subplot: large plot spanning two columns
plt.subplot(gs[0, :2])  
plt.plot(x, np.sin(x))
plt.title("Large Sine Wave")

# Second subplot: smaller plot
plt.subplot(gs[0, 2])
plt.plot(x, np.cos(x), 'r')
plt.title("Cosine")

# Third subplot: takes one row and two columns
plt.subplot(gs[1, :])
plt.plot(x, np.tan(x), 'g')
plt.title("Tangent")

# Fourth subplot: bottom right
plt.subplot(gs[2, 1:])
plt.plot(x, -np.sin(x), 'b')
plt.title("Negative Sine Wave")

plt.tight_layout()
plt.show()
```

**subplot() vs subplots():**
- subplot(): Used to create one subplot at a time. You need to call it multiple times to create multiple subplots.
- subplots(): A more modern and concise way to create subplots. It returns both the figure object and an array of axes, making it easier to handle multiple subplots.

**Key Points About Subplots:**
    
- Grid Layout: subplot() allows you to create a grid of subplots.
- Flexible Layouts: You can use GridSpec for more advanced and flexible layouts.
- Sharing Axes: You can share axes across subplots to make plots more coherent.
- Combining Plot Types: Subplots can be used to combine different plot types (e.g., line plots, scatter plots, bar charts) in a single figure.
- tight_layout(): Automatically adjusts the spacing between subplots to avoid overlap.
