<h4>Group data using custom function: Let's say you want to group your data using custom function. Here the requirement is to create three groups<h4>
<ol>
    <li>Days when temperature was between 80 and 90</li>
    <li>Days when it was between 50 and 60</li>
    <li>Days when it was anything else</li>
</ol>

```python
# For this you need to write custom grouping function and pass that to groupby

def grouper(df, idx, col):
    if 80 <= df[col].loc[idx] <= 90:
        return '80-90'
    elif 50 <= df[col].loc[idx] <= 60:
        return '50-60'
    else:
        return 'others'
```

```python
g = df.groupby(lambda x: grouper(df, x, 'temperature'))
g
```

```python
for key, d in g:
    print("Group by Key: {}\n".format(key))
    print(d)
```

Since we know that data is not always correct. Let's say order of cities in `windspeed_df` is different(as compare to `temperature_df`) and having missing data for `banglore`:

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
})

windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
})
```

```python
df = pd.concat([temperature_df,windspeed_df],axis=1)
df
```

As we can see the above data is not correct. In first row there should be mumbai instead of delhi.

In order to do that we can use `index` argument:

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
}, index=[1,0])
windspeed_df
```

```python
df = pd.concat([temperature_df,windspeed_df],axis=1)
df
```

Now this working correctly.

**REMARK:** `index` is way to align rows from different dataframe while using concat operation.

### merge

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3 = pd.merge(df1, df2, on="city")
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city",how="inner") # default is inner
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="left")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="right")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer",indicator=True)
df3
```
