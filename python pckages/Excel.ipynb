# Basic Mathematical Functions

### SUM

The SUM function adds numbers together. It's one of the most commonly used functions in Excel and is used for quick addition of values in a range of cells.

**Syntax:** `=SUM(number1, [number2], ...)`

- number1: The first number, cell reference, or range to sum. (Required)
- number2, ...: Additional numbers, cell references, or ranges to sum. (Optional)

```python
=SUM(A1:A5)
=SUM(10, 20, 30)

# If a cell in the range contains text, Excel ignores it. For example:
# A1: 10, A2: "text", A3: 20
=SUM(A1:A3) → 30
```

**Note:** Empty cells are treated as 0 and do not affect the result.
Text is ignored (unless it's a number stored as text).

### SUMIF

The SUMIF function adds numbers in a range that meet a specified condition or criteria. It's useful when you need to filter and sum values based on specific conditions, such as summing sales for a particular region or category.

**Syntax:** `=SUMIF(criteria_range, criteria, [sum_range])`

- criteria_range: The range of cells to evaluate for the criteria. (Required)
- criteria: The condition or criteria used to determine which cells to sum. (Required)
- sum_range: The range of cells to sum if the criteria are met. If omitted, Excel sums the cells in the range. (Optional)

```python
A      |  B
--------|------
Fruit   | Sales
Apple   | 100
Banana  | 150
Apple   | 200
Orange  | 250

# To sum sales for "Apple":
=SUMIF(A2:A5, "Apple", B2:B5)

# If you want to sum values greater than 100 in the B column:
=SUMIF(B2:B5, ">100")

# Using a Cell Reference in Criteria If D1 contains the word "Banana," you can use:
=SUMIF(A2:A5, D1, B2:B5)

# Omitting sum_range If the range itself contains numbers and you want to sum only the cells that meet the criteria:
=SUMIF(B2:B5, ">100")
```

- `*` (asterisk): Matches any number of characters. Example: `=SUMIF(A2:A5, "A*")` will sum sales for "Apple."
- `?` (question mark): Matches a single character. Example: `=SUMIF(A2:A5, "?range")` will sum sales for "Orange."

- Criteria like `">100"`, `"<50"`, or `"=200"` must be enclosed in double quotes.
- If you use a cell reference for criteria, combine the operator and reference with `&`: `=SUMIF(B2:B5, ">" & D1)
`


```python
A         |  B
-----------|-----
Region     | Sales
North      | 500
South      | 300
East       | 400
West       | 200

# Sum all sales greater than a value specified in cell D1. Assume value in D1 is 350 :
=SUMIF(B2:B5, ">" & D1)
```

`">" & D1`: Combines the `>` operator with the value in cell `D1`. Excel interprets this as `">350"`.

**Why Use `&`?**
The `&` symbol allows you to dynamically use the value in `D1` instead of hardcoding the condition like `">350"`. If you update `D1` to `250`, the formula will automatically adjust to sum all sales greater than 250.

### SUMIFS

The SUMIFS function adds numbers in a range that meet multiple specified criteria. It’s like an enhanced version of SUMIF, allowing you to specify multiple conditions.

**Syntax:** `=SUMIFS(sum_range, criteria_range1, criteria1, [criteria_range2, criteria2], ...)`

- sum_range: The range of cells to sum. (Required)
- criteria_range1: The range of cells to evaluate against the first condition. (Required)
- criteria1: The condition to apply to the first range. (Required)
- criteria_range2, criteria2, ...: Additional ranges and their respective conditions. (Optional)

**Key Points:**
- All criteria must be satisfied for a cell in the sum range to be included in the total.
- The ranges for `sum_range` and `criteria_range` must have the same size, or the function will return an error.
- Criteria can include numbers, text, cell references, or logical operators.

```python
A       |  B         |  C
Region  | Product    | Sales
North   | Apple      | 100
South   | Banana     | 150
North   | Banana     | 200
South   | Apple      | 250

# Sum sales for "North" and "Apple."
=SUMIFS(C2:C5, A2:A5, "North", B2:B5, "Apple")

#---------------------------------------------------------

A       |  B
Region  | Sales
North   | 100
South   | 150
East    | 200
West    | 250

# Sum sales greater than 150 in the "South" region.
=SUMIFS(B2:B5, A2:A5, "South", B2:B5, ">150")

# If E1 contains "North" and F1 contains ">100":
=SUMIFS(C2:C5, A2:A5, E1, C2:C5, F1) # formula dynamically uses E1 and F1 as criteria for flexibility.
```

- Wildcards can be used in text-based criteria
- Text criteria are not case-sensitive.
- To include wildcards or operators (>, <), enclose them in quotes.

### AVERAGE

The AVERAGE function calculates the arithmetic mean (average) of a group of numbers. It sums up the numbers and divides the total by the count of the numbers.

**Syntax:** `=AVERAGE(number1, [number2], ...)`

- number1: The first number, cell reference, or range. (Required)
- number2, ...: Additional numbers, cell references, or ranges to include in the average. (Optional)

```python
=AVERAGE(A1:A5)
=AVERAGE(10, 20, 30)

# You can use named ranges or dynamic ranges for flexibility. For example, if a named range Sales represents A1:A5:
=AVERAGE(Sales)
```

### ROUND

The ROUND function rounds a number to a specified number of digits. It uses standard rounding rules:

    - If the digit to the right is 5 or greater, it rounds up.
    - Otherwise, it rounds down.

**Syntax:** `=ROUND(number, num_digits)`

- number: The number to round. (Required)
- num_digits: The number of digits to which the number is rounded. (Required)
    - 0: Rounds to the nearest integer.
    - Positive value: Rounds to the specified number of decimal places.
    - Negative value: Rounds to the left of the decimal point.

```python
=ROUND(12.3456, 2) # 12.35
=ROUND(123.456, 0) # 123
=ROUND(1456, -2) # 1500
```

### ROUNDUP

The ROUNDUP function always rounds numbers up (away from zero), regardless of the value of the digit to the right.

**Syntax:** `=ROUNDUP(number, num_digits)`

- Same parameters as the ROUND function.

```python
=ROUNDUP(12.3456, 2) # 12.35
=ROUNDUP(123.456, 0) # 124
=ROUNDUP(1456, -2) # 1500
```

### ROUNDDOWN 

The ROUNDDOWN function always rounds numbers down (toward zero), regardless of the value of the digit to the right.

**Syntax:** `=ROUNDDOWN(number, num_digits)`

- Same parameters as the ROUND function.

```python
=ROUNDDOWN(12.3456, 2) # 12.34
=ROUNDDOWN(123.456, 0) # 123
=ROUNDDOWN(1456, -2) # 1400
```

### COUNT

The COUNT function counts the number of numeric values in a range or list of arguments. It ignores text, blanks, and logical values.

**Syntax:** `=COUNT(value1, [value2], ...)`

- value1, value2, ...: The values or ranges to count numeric entries. (Required)

```python
=COUNT(A1:A5) 
# If A1:A5 contains {10, "text", 20, , 30}, the result is 3 (only numeric values are counted).

=COUNT(10, 20, "hello", TRUE) # 2
```

### COUNTA

The COUNTA function counts all non-blank cells, including numbers, text, and logical values, but excluding completely empty cells.

**Syntax:** `=COUNTA(value1, [value2], ...)`

- value1, value2, ...: The values or ranges to count non-blank entries. (Required)

```python
=COUNTA(A1:A5) 
# If A1:A5 contains {10, "text", , TRUE, 30}, the result is 4 (ignores the blank cell).

=COUNTA(10, "", "Hello", FALSE) # 3
```

### COUNTIF

The COUNTIF function counts the number of cells that meet a single condition in a range.

**Syntax:** `=COUNTIF(range, criteria)`

- range: The range of cells to count. (Required)
- criteria: The condition that determines which cells to count. (Required)

```python
=COUNTIF(A1:A5, ">10")
# If A1:A5 contains {5, 15, 10, 20, 25}, the result is 3 (15, 20, 25).

=COUNTIF(A1:A5, "Apple")
# If A1:A5 contains {Apple, Banana, Apple, Orange}, the result is 2.

=COUNTIF(A1:A5, B1)
# If B1 contains ">10", the result depends on the condition in B1.
```

### COUNTIFS

The COUNTIFS function counts the number of cells that meet multiple conditions in one or more ranges.

**Syntax:** `=COUNTIFS(criteria_range1, criteria1, [criteria_range2, criteria2], ...)`

- criteria_range1, criteria1, ...: The ranges and conditions to apply. (Required)

```python
A1:A5: {5, 15, 20, 30, 25}
B1:B5: {25, 15, 20, 35, 10}


=COUNTIFS(A1:A5, ">10", B1:B5, "<30") # 2 (15 and 20 in both ranges meet the conditions).

=COUNTIFS(A1:A5, "A*")
# If A1:A5 contains {Apple, Banana, Apricot, Orange}, the result is 2.

A       |  B
------- | -----
10      | 20
15      | 25
5       | 30
20      | 10

# Count numbers greater than 10:
=COUNTIF(A1:A4, ">10")

# Count non-blank cells:
=COUNTA(A1:A4)

# Count numbers in A1:A4 greater than 10 and numbers in B1:B4 less than 25:
=COUNTIFS(A1:A4, ">10", B1:B4, "<25")
```

**Syntax:** `=MIN(number1, [number2], ...)`

- number1, number2, ...: The numbers or ranges to evaluate. (Required)

```python
=MIN(A1:A5)
=MIN(10, 20, 30, 5, 25)
```

### MAX

The MAX function returns the largest value in a given set of numbers.

**Syntax:** `=MAX(number1, [number2], ...)`

- number1, number2, ...: The numbers or ranges to evaluate. (Required)

```python
=MAX(A1:A5)
=MAX(10, 20, 30, 5, 25)

=MAX(A1:A5) - MIN(A1:A5)
```

### ABS

The ABS function removes the sign of a number, returning its absolute value. Positive numbers remain unchanged, while negative numbers become positive.

**Syntax:** `=ABS(number)`

- number: The number for which you want the absolute value. (Required)

```python
=ABS(-10) # 10
=ABS(15) # 15
=ABS(A1)

=ABS(A1 - B1)
# If A1 = 20 and B1 = 50, the result is 30 (|20 - 50|).
```

### MEDIAN

The MEDIAN function finds the median (middle value) of a set of numbers. If there is an odd number of values, it returns the middle one; if even, it returns the average of the two middle numbers.

**Syntax:** =MEDIAN(number1, [number2], ...)`

- number1, number2, ...: The numbers or ranges to find the median. (Required)

```python
=MEDIAN(5, 10, 15) # 10
=MEDIAN(10, 20, 30, 40) # 25 (average of 20 and 30).
=MEDIAN(A1:A5)
```

**Key Points:**

- MEDIAN ignores non-numeric values, including text and blank cells.
- The median is less affected by outliers than the average (mean).
    
    Example: {10, 20, 30, 1000} → MEDIAN = 25, while AVERAGE = 265.

The CONCATENATE function (or its modern replacement, CONCAT) combines two or more strings or values into a single text string.

**Syntax:** `=CONCATENATE(text1, [text2], ...)`

- text1, text2, ...: The strings or cell references to join. (Required)

```python
=CONCATENATE("Hello", " ", "World") # Hello World
=CONCATENATE(A1, " ", B1)
```

Microsoft recommends using the `CONCAT` function instead of CONCATENATE because it is more flexible.

- text: The text or cell reference containing the text to format. (Required)

### TRIM

The TRIM function removes all extra spaces from a text string, leaving only single spaces between words. It’s particularly useful for cleaning up messy data imported from external sources.


**Syntax:** `=TRIM(text)`

- text: The text or cell reference containing the text to clean. (Required)


```python
=TRIM("   Hello   World   ") # Hello World
=TRIM(A1)

# Using TRIM with concatenated text: If A1 = " John " and B1 = " Doe ", then:
=TRIM(A1) & " " & TRIM(B1) # John Doe
```

### LEFT

The LEFT function extracts a specified number of characters from the beginning (left) of a text string.

**Syntax:** `=LEFT(text, [num_chars])`

- text: The text or cell reference to extract characters from. (Required)
- num_chars: The number of characters to extract from the left. (Optional; defaults to 1 if omitted.)

```python
=LEFT("Excel", 3) # Exc
=LEFT(A1, 5)

# Default behavior (1 character):
=LEFT("Excel") # E
```

### RIGHT

The RIGHT function extracts a specified number of characters from the end (right) of a text string.

**Syntax:** `=RIGHT(text, [num_chars])`

- text: The text or cell reference to extract characters from. (Required)
- num_chars: The number of characters to extract from the right. (Optional; defaults to 1 if omitted.)

```python
=RIGHT("Excel", 3) # cel

# If A1 = "Hello World", then:
=RIGHT(A1, 5) # World

# Default behavior (1 character):
=RIGHT("Excel") # L
```

### MID

The MID function extracts a specified number of characters from the middle of a text string, starting at a defined position.

**Syntax:** `=MID(text, start_num, num_chars)`

- ntext: The text or cell reference to extract characters from. (Required)
- start_num: The position of the first character to extract (1-based index). (Required)
- num_chars: The number of characters to extract. (Required)

```python
=MID("Excel", 2, 3) # xce

# If A1 = "Hello World", then:
=MID(A1, 7, 5) # World
```

**Key points:**

- If `num_chars` or `start_num` exceeds the length of the text, Excel will return the available characters without an error.
- If a cell contains a numeric value, Excel will treat it as text for these functions. Example:

```python
 =LEFT(12345, 3) # 123

# Combine these with LEN, FIND, or SEARCH for more dynamic results:
# Extract the first name from A1 = "John Doe":
=LEFT(A1, FIND(" ", A1) - 1) # John
```

### FIND

The FIND function locates the starting position of a specific character or substring within a text string. It is `case-sensitive` and `does not support wildcards`.

**Syntax:** `=FIND(find_text, within_text, [start_num])`

- find_text: The text to find. (Required)
- within_text: The text or cell reference to search within. (Required)
- start_num: The position to start searching from. (Optional; defaults to 1 if omitted.)

```python
=FIND("e", "Excel") # 2 (position of the first "e")
=FIND("cel", "Excel") # 3

# If A1 = "Hello World", then:
=FIND("o", A1, 6) # 8 (finds "o" starting from the 6th character)

=FIND("E", "Excel") # #VALUE! (uppercase "E" is not found)
```

### SEARCH

The SEARCH function also finds the position of a character or substring but is `not case-sensitive` and `supports wildcards (* and ?)`.

**Syntax:** `=SEARCH(find_text, within_text, [start_num])`

- find_text: The text to find. (Required)
- within_text: The text or cell reference to search within. (Required)
- start_num: The position to start searching from. (Optional; defaults to 1 if omitted.)

```python
=SEARCH("e", "Excel") # 2
=SEARCH("CEL", "Excel") # 3 (case-insensitive)

=SEARCH("e*", "Excel") # 2 (starts matching "e" and the rest of the text)

# If A1 = "Excel excel", then:
=SEARCH("o", A1, 6) # 8

# If A1 = "John Doe", you can extract the first name:
=LEFT(A1, FIND(" ", A1) - 1) # John

# If A1 = "user@example.com", extract the domain:
=MID(A1, FIND("@", A1) + 1, LEN(A1) - FIND("@", A1)) # example.com

# If A1 = "Excel is fun": (Checking if a Substring Exists)
=IF(ISNUMBER(SEARCH("fun", A1)), "Found", "Not Found") # Found
```

### SUBSTITUTE

he SUBSTITUTE function replaces all or specific occurrences of a substring within a text string.

**Syntax:** `=SUBSTITUTE(text, old_text, new_text, [instance_num])`

- text: The original text or cell reference. (Required)
- old_text: The text you want to replace. (Required)
- new_text: The text you want to replace it with. (Required)
- instance_num: The specific occurrence of old_text to replace. (Optional; if omitted, all occurrences are replaced.)

### REPLACE

The REPLACE function replaces a portion of a text string based on the starting position and length of characters.

- old_text: The original text or cell reference. (Required)
- start_num: The position of the first character to replace (1-based index). (Required)
- num_chars: The number of characters to replace. (Required)
- new_text: The text to insert in place of the replaced portion. (Required)

```python
=REPLACE("Hello World", 7, 5, "Excel") # Hello Excel
=REPLACE("123456", 3, 3, "ABC") # 12ABC

# If A1 = "Excel Functions", then:
=REPLACE(A1, 1, LEN(A1), "Replaced") # Replaced
```

# Logical Functions

### IF

The IF function performs a logical test and returns one value if the condition is TRUE and another if FALSE.

**Syntax:** `=IF(logical_test, value_if_true, value_if_false)`

- logical_test: The condition to evaluate. (Required)
- value_if_true: The value to return if the condition is TRUE. (Required)
- value_if_false: The value to return if the condition is FALSE. (Required)

```python
=IF(A1 > 50, "Pass", "Fail")

# Check if a cell is empty:
=IF(A1 = "", "Blank", "Not Blank")

# Nested IF (multiple conditions):
=IF(A1 > 90, "A", IF(A1 > 75, "B", "C"))
# If A1 = 95, Result: A
# If A1 = 80, Result: B
# If A1 = 60, Result: C
```

### IFS

The IFS function tests multiple conditions and returns the value corresponding to the first TRUE condition. It’s more concise than nested IF functions.

**Syntax:** `=IFS(logical_test1, value_if_true1, [logical_test2, value_if_true2], ...)`

- logical_test1: The first condition to evaluate. (Required)
- value_if_true1: The value to return if the first condition is TRUE. (Required)
- logical_test2, value_if_true2: Additional conditions and values. (Optional)

```python
=IFS(A1 >= 90, "A", A1 >= 80, "B", A1 >= 70, "C", A1 < 70, "Fail")
# If A1 = 85, Result: B
# If A1 = 65, Result: Fail

=IFS(A1 = "Red", "Stop", A1 = "Yellow", "Caution", A1 = "Green", "Go")
# If A1 = "Yellow", Result: Caution

=IFS(A1 > 0, "Positive", A1 < 0, "Negative", TRUE, "Zero")
# If A1 = 10, Result: Positive
# If A1 = -5, Result: Negative
# If A1 = 0, Result: Zero
```

**Best Practices:**

- Avoid overly complex nested IFs: Use IFS for better readability.
- Include default values in IFS: `Use TRUE as the final condition to handle unexpected inputs`.
- Test your formulas: Ensure all possible inputs are accounted for to avoid errors.

### AND

The AND function tests multiple conditions and returns TRUE if all conditions are TRUE. If any condition is FALSE, it returns FALSE.

**Syntax:** `=AND(logical1, [logical2], ...)`

- logical1, logical2, ...: Conditions to test. (At least one is required, up to 255 conditions.)

```python
=AND(A1 > 50, A1 < 100)
# If A1 = 75, Result: TRUE
# If A1 = 120, Result: FALSE

=IF(AND(B1 >= 60, C1 >= 60), "Pass", "Fail")
# If B1 = 65 and C1 = 70, Result: Pass
# If B1 = 50 and C1 = 70, Result: Fail

=AND(A1 > 0, A2 > 0, A3 > 0)
# If all values in A1, A2, and A3 are positive, Result: TRUE
```

### OR

The OR function tests multiple conditions and returns TRUE if at least one condition is TRUE. It returns FALSE only if all conditions are FALSE.

**Syntax:** `=OR(logical1, [logical2], ...)`

- logical1, logical2, ...: Conditions to test. (At least one is required, up to 255 conditions.)

```python
=OR(A1 > 50, A1 < 20)
# If A1 = 75, Result: TRUE
# If A1 = 40, Result: FALSE

=IF(OR(B1 >= 60, C1 >= 60), "Pass", "Fail")
# If B1 = 50 and C1 = 70, Result: Pass
# If B1 = 50 and C1 = 50, Result: Fail

=OR(A1 > 100, A2 > 100, A3 > 100)
# If any value in A1, A2, or A3 is greater than 100, Result: TRUE
```

### NOT

The NOT function reverses the logical value of a condition. If the condition is TRUE, NOT returns FALSE, and vice versa.

**Syntax:** `=NOT(logical)`

- logical: The condition to evaluate. (Required)

```python
=EOMONTH(TODAY(), 0) - TODAY() # Returns the number of days remaining until the end of the current month.

=EOMONTH(TODAY(), -MOD(MONTH(TODAY())-1, 3)) 
# This formula returns the last day of the previous quarter based on today's date.
```

### SORT

The SORT function is a powerful tool that allows you to sort a range or array of data dynamically. It enables sorting based on columns or rows, in ascending or descending order.

- Returns a sorted array of data based on specified criteria.
- Can sort data by one or multiple columns/rows.
- Automatically updates when the source data changes.


**Syntax:** `=SORT(array, [sort_index], [sort_order], [by_col])`

- array: The range or array to sort. (Required)
- sort_index: The column (or row) number to sort by within the array. (Optional; default is 1.)
- sort_order: The order to sort:
    - 1 for ascending (default).
    - -1 for descending.
- by_col: Logical value indicating whether to sort by columns (TRUE) or rows (FALSE). (Optional; default is FALSE.)

```python
=SORT(A1:A10) 
# Sorted list with the smallest value at the top.

=SORT(A1:C10, 2)
# Sorts the range A1:C10 based on the second column in ascending order.
 
=SORT(A1:C10, 2, -1)
# Sorts the range A1:C10 by the second column in descending order.

=SORT(A1:C10, {2, 1}, {1, -1})
# Sorts the range A1:C10 first by the second column (ascending) and then by the first column (descending).


=SORT(A1:E1, 1, 1, TRUE)
# Sorts the row range A1:E1 in ascending order.
```

**Note:** If the result of SORT overlaps an existing range, you'll see a #SPILL! error. Ensure the output range is clear.

```python
=SORT(FILTER(A1:C10, B1:B10>50), 3, -1)

=SORT(UNIQUE(A1:A20)) 
# Removes duplicates from A1:A20 and sorts the unique values.
```

### SORTBY

The SORTBY function allows you to sort a range or array based on values in a related array or column. It's more flexible than the SORT function because you can specify multiple sorting criteria across different arrays.

- Dynamically sorts a range or array based on one or more sorting arrays.
- Can handle sorting by arrays not included in the main data range.
- Allows for multi-level sorting in ascending or descending order.


**Syntax:** `=SORTBY(array, by_array1, [sort_order1], [by_array2], [sort_order2], ...)`

- array: The range or array to sort. (Required)
- by_array1: The first array or range to sort by. (Required)
- sort_order1: The sort order for by_array1:
    - 1 for ascending (default).
    - -1 for descending.
- by_array2, sort_order2: (Optional) Additional arrays and sort orders for secondary or further sorting.

```python
=SORTBY(A1:B10, B1:B10, 1)
# Sorts the range A1:B10 by the values in column B in ascending order.

=SORTBY(A1:C10, B1:B10, 1, C1:C10, -1)
# Sorts the range A1:C10:
    # By column B in ascending order.
    # If values in B are equal, sorts by column C in descending order.

# If you have a dataset in A1:C10 and want to dynamically sort it based on a user-defined column in cell E1:
=SORTBY(A1:C10, INDEX(A1:C10, 0, E1), 1) # ormula sorts by the column number specified in E1

=SORTBY(A1:A10, D1:D10, -1)
# Sorts the range A1:A10 based on the values in D1:D10 in descending order.

# If you have months as text (e.g., January, February) and want to sort them in chronological order:
=SORTBY(A1:A12, MONTH(DATEVALUE(A1:A12 & " 1")), 1)

=SORTBY(A1:C10, B1:B10, 1, C1:C10, -1)
# First sorts by B1:B10 (ascending), then by C1:C10 (descending).

=SORTBY(FILTER(A1:C10, C1:C10>50), B1:B10, -1)
# Filters rows where column C > 50, then sorts by column B in descending order.
```

### FILTER

The FILTER function extracts and returns a subset of data that meets specified criteria. It’s one of the most powerful dynamic array functions in Excel, allowing you to filter data dynamically without altering the original dataset.

- Returns a filtered array of data based on a condition.
- The result updates automatically when the source data or criteria change.

**Syntax:** `=FILTER(array, include, [if_empty])`

- array: The range or array to filter. (Required)
- include: A logical array (criteria) determining which rows/columns to include. (Required)
- if_empty: The value to return if no data meets the criteria. (Optional; default is #CALC!.)

```python
=FILTER(A1:C10, B1:B10>50)
# Filters rows from A1:C10 where the values in column B are greater than 50.

=FILTER(A1:C10, B1:B10>50, "No records found")
# Returns "No records found" if no rows match the condition.

=FILTER(A1:C10, (B1:B10>50)*(C1:C10<100)) # Filter by Multiple Criteria (AND Logic)
# Includes rows where column B > 50 and column C < 100.

=FILTER(A1:C10, (B1:B10>50)+(C1:C10<100)) # Filter by Multiple Criteria (OR Logic)
#Includes rows where column B > 50 or column C < 100.

=FILTER(A1:C10, B1:B10=E1)
# Filters rows where column B matches the value in cell E1.

=FILTER(A1:C10, B1:B10="Sales")
# Filters rows where column B contains the text "Sales".

=FILTER(A1:C10, ISNUMBER(SEARCH("abc", B1:B10)))
# Filters rows where column B contains the substring "abc".

=FILTER(A1:C10, A1:A10>TODAY())
# Filters rows where column A contains dates later than today.

=SORT(FILTER(A1:C10, B1:B10>50), 2, 1)
# Filters rows where column B > 50 and sorts the result by the second column in ascending order.
```

**Keep note:**

- Always include the if_empty argument to handle cases where no data matches the criteria.
- Use `*` for AND logic and `+` for OR logic.
- The result of FILTER spills into adjacent cells. Ensure the output range is clear to avoid #SPILL! errors.

### UNIQUE

The UNIQUE function returns a list of unique values from a range or array. It dynamically eliminates duplicates, making it a powerful tool for data analysis and organizing datasets.

- Extracts unique values from a range or array.
- Can return distinct values for rows, columns, or both.
- Dynamically updates when the source data changes.


**Syntax:** `=UNIQUE(array, [by_col], [exactly_once])`

- array: The range or array from which to extract unique values. (Required)
- by_col: Indicates whether to compare values by column (TRUE) or row (FALSE). Defaults to FALSE. (Optional)
- exactly_once: Determines whether to return only values that appear exactly once (TRUE) or all unique values (FALSE). Defaults to FALSE. (Optional)

```python
=UNIQUE(A1:A10)
# Returns a list of unique values from the range A1:A10.

=UNIQUE(A1:C10)
# Returns unique rows from the range A1:C10.

=UNIQUE(A1:D5, TRUE)
# Extracts unique values by comparing columns in the range A1:D5.

=UNIQUE(A1:A10, FALSE, TRUE)
# Returns values from A1:A10 that appear exactly once in the range.

=SORT(UNIQUE(A1:A10))
# Extracts unique values from A1:A10 and sorts them in ascending order.

=UNIQUE(FILTER(A1:A10, B1:B10>50))
# Filters A1:A10 where B1:B10 > 50, then returns unique values from the filtered range.

=COUNTA(UNIQUE(A1:A10))
# Counts the number of unique values in A1:A10.

=UNIQUE(FILTER(A1:A10, ISTEXT(A1:A10)))
# Returns unique text values from the range A1:A10.

=UNIQUE(A1:A10)
# Works seamlessly to extract unique dates from A1:A10.
```

**Important Notes:**

- The UNIQUE function is `not case-sensitive`. For case-sensitive unique extraction, you would need additional formulas or helper columns.
- Often combined with functions like FILTER, SORT, or COUNTIF for advanced data manipulation.

### TAKE

The TAKE function is used to return a specified number of rows or columns from the start or end of an array. It allows for slicing arrays dynamically to extract specific sections.

- Extracts a specific number of rows or columns from an array.
- Works dynamically, updating results if the source data changes.


**Syntax:** `=TAKE(array, [rows], [columns])`

- array: The range or array from which to extract rows or columns. (Required)
- rows: The number of rows to return. A positive value extracts rows from the start, and a negative value extracts rows from the end. (Optional)
- columns: The number of columns to return. A positive value extracts columns from the start, and a negative value extracts columns from the end. (Optional)

```python
=TAKE(A1:C10, 3)
# Returns the first 3 rows of the range A1:C10.

=TAKE(A1:C10, -3)
# Returns the last 3 rows of the range A1:C10.

=TAKE(A1:C10, , 2)
# Returns the first 2 columns of the range A1:C10.

=TAKE(A1:C10, , -2)
# Returns the last 2 columns of the range A1:C10.

=TAKE(A1:C10, 3, 2)
# Returns the first 3 rows and first 2 columns from the range A1:C10.

=TAKE(SORT(A1:C10, 2, -1), 5)
# Sorts A1:C10 by the second column in descending order and returns the top 5 rows.

=TAKE(A1:C10, E1)
# Returns the number of rows specified in E1.

=TAKE(FILTER(A1:C10, B1:B10>50), 3)
# Filters rows where column B > 50 and then takes the first 3 rows of the result.

=TAKE(DROP(A1:C10, 3), 4)
# Drops the first 3 rows and then takes the next 4 rows.

```

**Syntax:** `=DROP(array, [rows], [columns])`

- array: The range or array from which to exclude rows or columns. (Required)
- rows: The number of rows to exclude. A positive value excludes rows from the start, and a negative value excludes rows from the end. (Optional)
- columns: The number of columns to exclude. A positive value excludes columns from the start, and a negative value excludes columns from the end. (Optional)

```python
=DROP(A1:C10, 3)
# Removes the first 3 rows from the range A1:C10.

=DROP(A1:C10, -3)
# Removes the last 3 rows from the range A1:C10.

=DROP(A1:C10, , 2)
# Removes the first 2 columns from the range A1:C10.

=DROP(A1:C10, , -1)
# Removes the last column from the range A1:C10.

=DROP(A1:C10, 2, 1)
# Removes the first 2 rows and the first column from the range A1:C10.

=TAKE(DROP(A1:C10, 2), 4)
# Removes the first 2 rows and then extracts the next 4 rows from the result.

=DROP(A1:C10, E1)
# Drops the number of rows specified in cell E1.

=DROP(SORT(A1:C10, 2, -1), 3)
# Sorts A1:C10 by the second column in descending order and removes the top 3 rows.

=DROP(A1:C10, , COUNTA(A1:C1)-2)
# Keeps only the first 2 columns dynamically, assuming COUNTA(A1:C1) counts the non-empty columns.
```

### VLOOKUP

The VLOOKUP function is one of Excel's most popular lookup functions. It searches for a value in the first column of a table and returns a value in the same row from a specified column.

- Stands for Vertical Lookup.
- Searches for a value vertically in the first column of a range or table.
- Returns a corresponding value from another column in the same row.

**Syntax:** `=VLOOKUP(lookup_value, table_array, col_index_num, [range_lookup])`

- lookup_value: The value to search for in the first column of the table. (Required)
- table_array: The range or table where the lookup is performed. The first column must contain the lookup values. (Required)
- col_index_num: The column number (starting from 1) in the table from which to return the value. (Required)
- range_lookup: Determines if the lookup should be an exact match (FALSE) or approximate match (TRUE). (Optional, default is TRUE.)

**How It Works**
- Searches for the lookup_value in the first column of table_array.
- Finds the closest match based on range_lookup.
- Returns the value from the column specified by col_index_num.                                                                                           

```python
A	B
Product	Price
Apple	2.00
Banana	1.50
Orange	1.80

=VLOOKUP("Banana", A2:B4, 2, FALSE) # 1.50
```

```python
A	B
Score	Grade
90	A
80	B
70	C
60	D

=VLOOKUP(75, A2:B5, 2, TRUE) # C
# Function looks for the largest value ≤ 75, which is 70.
```

```python
=VLOOKUP(D1, A2:B5, 2, FALSE)
# Here, D1 contains the lookup value.
# Example: If D1 = Orange, the formula will return 1.80.
```

```python
A	B	C
ID	Name	Department
101	Alice	HR
102	Bob	IT
103	Charlie	Finance

=VLOOKUP(102, A2:C4, 3, FALSE) # IT
```

**Common Errors**
  
- #N/A Error
    - Occurs when the lookup value is not found in the first column.
    - Fix: Check if the value exists and matches exactly.

- #REF! Error
    - Occurs when col_index_num exceeds the number of columns in table_array.
    - Fix: Ensure col_index_num is within the valid range.

- #VALUE! Error
    - Occurs when col_index_num is non-numeric.
    - Fix: Use a numeric column index

**Key notes:**

- The lookup value must always be in the first column of table_array.
- Use TRUE (default) for an approximate match in sorted data.
- Use FALSE for an exact match.
- VLOOKUP is not case-sensitive (e.g., Apple and apple are treated the same).
- Use INDEX-MATCH or XLOOKUP for more flexibility (e.g., backward lookup).

```python
=VLOOKUP("Bob", A2:C4, MATCH("Department", A1:C1, 0), FALSE)
# Dynamically determines the column index for "Department."
```

### HLOOKUP

The HLOOKUP function is the horizontal counterpart of VLOOKUP. It searches for a value in the first row of a table and returns a value in the same column from a specified row.

- Stands for Horizontal Lookup.
- Searches for a value in the first row of a range or table.
- Returns a corresponding value from another row in the same column.

**Syntax:** `=HLOOKUP(lookup_value, table_array, row_index_num, [range_lookup])`

- lookup_value: The value to search for in the first row of the table. (Required)
- table_array: The range or table where the lookup is performed. The first row must contain the lookup values. (Required)
- row_index_num: The row number (starting from 1) in the table from which to return the value. (Required)
- range_lookup: Determines if the lookup should be an exact match (FALSE) or approximate match (TRUE). (Optional, default is TRUE.)

**How It Works**
    
- Searches for the lookup_value in the first row of the table_array.
- Finds the closest match based on range_lookup.
- Returns the value from the row specified by row_index_num.                                                                                      

```python
A	B	C	D
Product	Apple	Banana	Orange
Price	2.00	1.50	1.80
Stock	100	200	150

=HLOOKUP("Banana", A1:D3, 2, FALSE) # 1.50
# Searches for "Banana" in the first row and returns the value in the second row of the same column.
```

```python
A	B	C	D	E
Score	60	70	80	90
Grade	D	C	B	A

=HLOOKUP(75, A1:E2, 2, TRUE) # C
# Finds the largest value ≤ 75 in the first row (which is 70) and 
# returns the value from the second row in the same column.
```

```python
=HLOOKUP(B1, A1:D3, 3, FALSE)
# If B1 = Orange, the formula returns the stock for Orange (150).
```

```python
A	B	C	D
Year	2020	2021	2022
Revenue	50000	60000	70000
Expenses	20000	25000	30000

=HLOOKUP(2021, A1:D3, 3, FALSE) # 25000
```

**Common Errors**
  
- #N/A Error
    - Occurs when the lookup value is not found in the first row.
    - Fix: Ensure the value exists and matches exactly.

- #REF! Error
    - Occurs when row_index_num exceeds the number of rows in table_array.
    - Fix: Use a valid row index.

- #VALUE! Error
    - Occurs when row_index_num is non-numeric.
    - Fix: Use a numeric row index.

**Key notes:**
- The lookup value must always be in the first row of table_array.
- Use TRUE for an approximate match in sorted data.
- Use FALSE for an exact match.
- HLOOKUP is not case-sensitive (e.g., Banana and banana are treated the same).
- Consider using INDEX-MATCH or XLOOKUP for more flexible lookups.


```python
=HLOOKUP("Revenue", A1:D3, MATCH("2021", A1:A3, 0), FALSE)
# Dynamically determines the row for "2021."
```

### LOOKUP
he LOOKUP function searches for a value in one row or column and returns a corresponding value from another row or column. It is a simpler version of VLOOKUP or HLOOKUP but has limitations.

**Vector Form (Most Common) Syntax:** `=LOOKUP(lookup_value, lookup_vector, result_vector)`

- lookup_value → The value to search for.
- lookup_vector → A single row or column containing the search values (must be sorted in ascending order).
- result_vector → A single row or column containing the corresponding results.

✅ Best used when searching in a single column or row.

**Array Form (Less Common) Syntax:** `=LOOKUP(lookup_value, array)`

- lookup_value → The value to find.
- array → A 2D table where LOOKUP searches the first column/row and returns the last column/row value in that row.

⚠️ Array form is rarely used because it works only when the first column or row is sorted in ascending order.

```python
# Vector Form (Column Lookup)

ID (A)	Name (B)
101	Alice
102	Bob
103	Charlie
104	David

# Find the name for ID 103
=LOOKUP(103, A1:A5, B1:B5) # "Charlie"
```

```python
# Vector Form (Row Lookup)

Product (A1:E1)	A	B	C	D
Price (A2:E2)	10	20	30	40

# Find the price for Product C
=LOOKUP("C", A1:E1, A2:E2) # 30
```

```python
#  Array Form (Find Closest Match)

Score (A)	Grade (B)
50	F
60	D
70	C
80	B
90	A

# Find the grade for Score 75
=LOOKUP(75, A1:A5, B1:B5) #  "C"
```

1. LOOKUP always finds the closest lower match:
    - If an exact match isn't found, it returns the next smallest value.
    - The lookup_vector must be sorted in ascending order.
2. LOOKUP doesn't support an exact match like VLOOKUP/XLOOKUP:
    - If the lookup value is smaller than the smallest value in the array, it returns an error.
3. LOOKUP works only in one direction:
    - It only searches from left to right (unlike XLOOKUP, which works in any direction).

### XLOOKUP

The XLOOKUP function is a modern, versatile lookup function introduced in Excel 365 and Excel 2021. It is designed to overcome the limitations of older lookup functions like VLOOKUP and HLOOKUP.

- Searches for a value in a range or array and returns a corresponding value from another range or array.
- Can search both vertically and horizontally.
- Allows for exact, approximate, and even reverse lookups.    

**Syntax:** `=XLOOKUP(lookup_value, lookup_array, return_array, [if_not_found], [match_mode], [search_mode])`

- lookup_value: The value to search for. (Required)
- lookup_array: The range where the lookup value is searched. (Required)
- return_array: The range or array from which to return a result. Must be the same size as lookup_array. (Required)
- if_not_found: Value to return if lookup_value is not found. (Optional)
- match_mode:
    - 0 (Default): Exact match.
    - -1: Exact match or next smaller item.
    - 1: Exact match or next larger item.
    - 2: Wildcard match (*, ?, ~). (Optional)
- search_mode:
    - 1 (Default): Search from first to last.
    - -1: Search from last to first.
    - 2: Binary search (sorted ascending).
    - -2: Binary search (sorted descending). (Optional)


**How It Works**
    
- Searches the lookup_value in the lookup_array.
- Finds the match based on match_mode and search_mode.
- Returns the corresponding value from the return_array.                                                                              

```python
A	B
Product	Price
Apple	2.00
Banana	1.50
Orange	1.80

=XLOOKUP("Banana", A2:A4, B2:B4) # 1.50
# Searches for "Banana" in column A and returns the corresponding price from column B.
```

```python
A	B
Price	Product
2.00	Apple
1.50	Banana
1.80	Orange

=XLOOKUP(1.80, A2:A4, B2:B4) # Orange
# Searches for 1.80 in column A and returns the corresponding product from column B.
```

```python
=XLOOKUP("Grapes", A2:A4, B2:B4, "Not Found") # Not Found
# Returns "Not Found" if "Grapes" is not in the lookup array.
```

```python
A	B
Score	Grade
90	A
80	B
70	C
60	D

=XLOOKUP(75, A2:A5, B2:B5, "No Grade", -1) # C
# Finds the largest value ≤ 75, which is 70.
```

```python
Product	Apple	Banana	Orange
Price	2.00	1.50	1.80

=XLOOKUP("Banana", B1:D1, B2:D2) # 1.50
```

```python
=XLOOKUP("B*", A2:A4, B2:B4, "Not Found", 2) # 1.50
# Matches any product starting with "B".
```

**Advantages of XLOOKUP**

- Can look both forward and backward, unlike VLOOKUP and HLOOKUP.
- Unlike VLOOKUP, the default match mode is exact.
- Allows specifying a value (if_not_found) for unmatched lookups.
- Supports returning entire rows or columns when used with dynamic ranges.
- Combines the power of INDEX-MATCH in a simpler syntax.

**Key Notes**
- XLOOKUP is not case-sensitive.
- If the return array contains multiple values, XLOOKUP can spill results into adjacent cells.
- Faster and more versatile compared to older lookup functions.

```python
A	B	C	D
Product	Price	Stock	Category
Apple	2.00	100	Fruit
Banana	1.50	200	Fruit
Orange	1.80	150	Fruit

=XLOOKUP("Banana", A2:A4, B2:D4)
```

### MATCH

The MATCH function is used to locate the relative position of a value in a row, column, or table. It is often combined with other functions like INDEX for more advanced lookups.

- Returns the position of a value within a range (row or column), rather than the value itself.

**Syntax:** `=MATCH(lookup_value, lookup_array, [match_type])`

- lookup_value: The value you want to find. (Required)
- lookup_array: The range of cells to search. Must be a single row or column. (Required)
- match_type:
    - 1 (Default): Finds the largest value less than or equal to lookup_value. Requires the data to be sorted in ascending order.
    - 0: Finds an exact match.
    - -1: Finds the smallest value greater than or equal to lookup_value. Requires the data to be sorted in descending order.


**How It Works**
    
- MATCH searches the lookup_value in the lookup_array.
- Depending on the match_type, it returns:
    - The position of the value in the array if found.
    - An error (#N/A) if no match is found.                                                                   

```python
A	B
1	Apple
2	Banana
3	Orange

=MATCH("Banana", B1:B3, 0) # 2
# "Banana" is the second item in the range.
```

```python
A
10
20
30

=MATCH(25, A1:A3, 1) # 2
# Finds the largest value ≤ 25, which is 20, at position 2.
```

```python
A
50
40
30

=MATCH(35, A1:A3, -1) # 2
# Finds the smallest value ≥ 35, which is 40, at position 2.
```

```python
A	B
101	Apple
102	Banana
103	Orange

=INDEX(B1:B3, MATCH(102, A1:A3, 0)) # Banana
# inds 102 in column A, determines its position (2), and returns the value from the second row of column B.
```

```python
=MATCH(3, {1,2,3,4,5}, 0) # 3 (The position of 3 in the array is 3.)

# If the value isn't found, MATCH returns #N/A. You can handle this with the IFERROR function:
=IFERROR(MATCH(6, {1,2,3,4,5}, 0), "Not Found")
```

**Key Notes**

- MATCH is not case-sensitive. To perform a case-sensitive match, use INDEX with an array formula or VBA.
- lookup_array must be a single row or column.
- If you use a multi-column or multi-row range, MATCH will throw an error.
- When lookup_value isn’t found, MATCH returns #N/A.
- For match_type = 1, the data must be in ascending order.
- For match_type = -1, the data must be in descending order.
- Use `*` to match any number of characters.
- Use `?` to match a single character.
- This works only with match_type = 0.


### INDEX

The INDEX function is used to return the value of a cell in a range based on its row and column numbers. It is extremely useful for retrieving data from a specific position in a dataset.

- Returns the value of a cell or range of cells at a specified location within a table or array.

There are two forms of the INDEX function: Array Form and Reference Form.

**Array Form Syntax:** `=INDEX(array, row_num, [column_num])`

- array: The range of cells or array from which to retrieve the value. (Required)
- row_num: The row number in the array from which to retrieve the value. (Required)
- column_num: The column number in the array from which to retrieve the value. (Optional; defaults to 1 if omitted).

**Reference Form Syntax:** `=INDEX(reference, row_num, [column_num], [area_num])`

- reference: A reference to one or more cell ranges.
- row_num: The row number in the reference from which to retrieve the value.
- column_num: The column number in the reference from which to retrieve the value. (Optional)
- area_num: Specifies which range to use when reference contains multiple ranges. (Optional)


**How It Works**
    
- Excel retrieves the value based on the row and column numbers provided.
- If column_num is omitted, Excel assumes it is 1.
- If the value is outside the range, Excel returns an error (#REF!).                                                                

```python
A	B	C
Apple	Banana	Cherry
Orange	Mango	Peach

=INDEX(A1:C2, 2, 2) # Mango
# Retrieves the value from the 2nd row and 2nd column of the range.
```

```python
A	B
101	Apple
102	Banana
103	Cherry

=INDEX(B1:B3, MATCH(102, A1:A3, 0)) # Banana
# MATCH(102, A1:A3, 0) returns 2 (position of 102 in column A).
# INDEX(B1:B3, 2) retrieves the value in the 2nd row of column B (Banana).
```

```python
A	B	C
1	2	3
4	5	6
7	8	9

=INDEX(A1:C3, 2, 0) # {4, 5, 6}
# Returns all values in the 2nd row.

=INDEX(A1:C3, 0, 3) # {3, 6, 9}
# Returns all values in the 3rd column.
```

**Key Notes**

- If row_num or column_num exceeds the size of the range, Excel returns #REF!.
- Use 0 for either row_num or column_num to retrieve entire rows or columns.
- When combined with MATCH, it eliminates the limitations of VLOOKUP (e.g., left-to-right restriction).
- Works well in scenarios requiring dynamic or complex references.



```python
ID	Name	Age	Department	Salary
101	Alice	25	HR	50000
102	Bob	30	IT	60000
103	Charlie	35	Finance	70000
104	David	40	Marketing	80000
105	Eve	45	Sales	90000
```

### OFFSET

The OFFSET function is used to create a dynamic range by returning a reference to a range that is offset from a specified starting point by a certain number of rows and columns.

- Returns a reference to a range of cells that is a specific number of rows and columns away from a starting cell or range. It can also specify the height and width of the returned range.

**Syntax:** `=OFFSET(reference, rows, cols, [height], [width])`

- reference: The starting point for the range (single cell or range). (Required)
- rows: The number of rows to move from the reference. Positive numbers move down; negative numbers move up. (Required)
- cols: The number of columns to move from the reference. Positive numbers move to the right; negative numbers move to the left. (Required)
- height: The number of rows in the returned range. Defaults to 1 if omitted. (Optional)
- width: The number of columns in the returned range. Defaults to 1 if omitted. (Optional)


**How It Works**
    
- The function starts at the reference cell or range.
- Moves by rows and cols.
- Returns a range with the specified height and width.
                                                           

```python
A	B	C
10	20	30
40	50	60
70	80	90

=OFFSET(A1, 1, 1) # 50
# Starts at A1, moves down 1 row and right 1 column to B2.
```

```python
=OFFSET(A1, 1, 0, 2, 2) # {40, 50; 70, 80}
# Starts at A1, moves down 1 row, and returns a range 2 rows tall and 2 columns wide (range A2:B3).
```

- OFFSET is volatile, meaning it recalculates every time the worksheet changes, even if unrelated. This can slow performance in large workbooks.
- Used in dynamic ranges for charts or reports.
- Helps in creating flexible formulas based on changing data.
- Often used with SUM, AVERAGE, COUNTA, and other aggregation functions.

### STDEV.P and STDEV.S

The STDEV.P and STDEV.S functions in Excel are used to calculate the standard deviation of a dataset. Standard deviation measures the amount of variation or dispersion from the average (mean).

- `STDEV.P`: Calculates standard deviation for the entire population.
- `STDEV.S`: Calculates standard deviation for a sample of the population.

**When to Use STDEV.P vs. STDEV.S**
- STDEV.P (Population): Use when you have data for the entire population (all possible data points).
- STDEV.S (Sample): Use when you have data for a sample of the population (a subset of data points).

**Syntax:** `=STDEV.P(number1, [number2], ...)`

**Syntax:** `=STDEV.S(number1, [number2], ...)`

- number1, [number2], ...: Numeric values, ranges, or references containing numbers to calculate the standard deviation. (At least one value is required.)


**How It Works**
    
1. The function calculates the mean (average) of the data.
2. It determines the squared difference of each data point from the mean.
3. It calculates the average of these squared differences.
4. For:
    - STDEV.P: Divides by the total number of data points.
    - STDEV.S: Divides by one less than the number of data points (n - 1).
    

```python
A
5
10
15
20

=STDEV.P(A1:A4) # 5.59
# Standard deviation for all data points.

=STDEV.S(A1:A4) # 6.45
# Standard deviation for a sample of data points.
```

```python
=STDEV.P(5, 10, 15, 20) # 5.59
=STDEV.S(5, 10, 15, 20) # 6.45

# If the data is in cells B1:B10:
=STDEV.P(B1:B10)
=STDEV.S(B1:B10)
```

```python
A
2
4
6
8
10

=STDEV.P(A1:A5) # 2.83
=STDEV.S(A1:A5) # 3.16

# Key Difference:

# STDEV.P divides by 5 (total data points).
# STDEV.S divides by 4 (n - 1, where n = 5).
```

**Important Notes:**

- STDEV.P and STDEV.S ignore text, logical values, and empty cells in ranges.
- Use constants like TRUE (evaluates as 1) or FALSE (evaluates as 0) in formulas if needed.
- If all arguments are non-numeric or invalid, Excel returns #DIV/0!.
- Useful for analyzing large datasets, especially for statistical summaries.
- Use VAR.P and VAR.S to calculate variance if needed (square of standard deviation).

### VAR.P and VAR.S

The VAR.P and VAR.S functions calculate the variance of a dataset, a statistical measure of how data points differ from the mean. Variance is the square of the standard deviation.

- VAR.P: Computes variance for the entire population.
- VAR.S: Computes variance for a sample of the population

**When to Use STDEV.P vs. STDEV.S**
- VAR.P (Population Variance): Use when you have data for the entire population (all possible data points).
- VAR.S (Sample Variance): Use when you have data for a sample of the population (a subset of data points).

**Syntax:** `=VAR.P(number1, [number2], ...)`

**Syntax:** `=VAR.S(number1, [number2], ...)`

- number1, [number2], ...: Numeric values, ranges, or references containing numbers for variance calculation. (At least one value is required.)


**How It Works**
    
1. The function calculates the mean (average) of the data.
2. It determines the squared difference of each data point from the mean.
3. For:
    - VAR.P: Divides the total squared differences by the total number of data points.
    - VAR.S: Divides the total squared differences by one less than the number of data points (n - 1).
    

```python
A
5
10
15
20

=VAR.P(A1:A4) # 31.25
# Variance for all data points is computed by dividing the squared differences by 4.

=VAR.S(A1:A4) # 41.67
# Variance for the sample is computed by dividing the squared differences by (4 - 1).
```

```python
=VAR.P(5, 10, 15, 20) # 31.25
=VAR.S(5, 10, 15, 20) # 41.67

# If the data is in cells B1:B10:
=VAR.P(B1:B10)
=VAR.S(B1:B10)
```

```python
A
2
4
6
8
10

=VAR.P(A1:A5) # 8
=VAR.S(A1:A5) # 10

# Key Difference:

# VAR.P divides by 5 (total data points).
# VAR.S divides by 4 (n - 1, where n = 5).
```

**Practical Use Cases**

- Variance helps assess the risk of investments by measuring how much returns deviate from the average.
- In manufacturing, variance can indicate inconsistencies in production processes.
- Useful in hypothesis testing or analyzing variability in sample data.

### CORREL

The CORREL function calculates the correlation coefficient between two datasets. The correlation coefficient is a statistical measure that shows the strength and direction of the linear relationship between two variables.

**Key Points about the CORREL Function**
    
1. Output:
    - A value between -1 and 1:
        - 1: Perfect positive correlation.
        - -1: Perfect negative correlation.
        - 0: No correlation.

2. Use Case:
- To determine if two datasets are related and how strongly they are related (e.g., sales vs. advertising spend, temperature vs. ice cream sales).


**Syntax:** `=CORREL(array1, array2)`

- array1: The first range of data (independent variable or x).
- array2: The second range of data (dependent variable or y).

**Rules and Requirements**

- Both arrays must have the same number of data points.
- Both arrays must contain numeric values.
- Blank cells, text, or logical values are ignored.
- If the standard deviation of either array is 0, Excel returns the #DIV/0! error.

**How It Works**

- Calculates the average of each dataset.
- Determines the covariance between the datasets.
- Divides the covariance by the product of the standard deviations of the datasets.


```python
# You want to determine the relationship between study hours and test scores.

A (Study Hours)	B (Test Scores)
2	50
4	70
6	80
8	90
10	95

=CORREL(A1:A5, B1:B5) # 0.986
# There is a strong positive correlation between study hours and test scores.
```

```python
Advertising Spend ($)	Sales ($)
500	3000
700	4000
900	5000
1100	6000
1300	7000

=CORREL(A1:A5, B1:B5) # 1.0
# Perfect positive correlation; as advertising spend increases, sales increase proportionally.
```

**Error Handling**
  
- #N/A: Occurs if the arrays have different lengths.
- #DIV/0!: Occurs if the standard deviation of either array is zero or all data points in an array are identical.

**Visualization Tip**

### COVARIANCE.P and COVARIANCE.S

The COVARIANCE.P and COVARIANCE.S functions in Excel are used to calculate covariance, which measures how two datasets vary together.

**Differences Between COVARIANCE.P and COVARIANCE.S**
1. COVARIANCE.P:
    - Calculates covariance for an entire population.
2. COVARIANCE.S:
    - Calculates covariance for a sample of the population.

The distinction lies in the denominator:

- COVARIANCE.P divides by n (total number of data points).
- COVARIANCE.S divides by n−1 (degrees of freedom).



**Syntax:** `=COVARIANCE.P(array1, array2)`

**Syntax:** `=COVARIANCE.S(array1, array2)`

- array1: The first dataset (e.g., independent variable).
- array2: The second dataset (e.g., dependent variable).


**Rules and Requirements**

1. Both arrays must have the same number of elements.

### FREQUENCY

The FREQUENCY function is a powerful statistical tool that calculates how often values occur within specified ranges, or bins. It is particularly useful for summarizing data distributions.


**Syntax:** `=FREQUENCY(data_array, bins_array)`

- data_array: The range of numeric values you want to analyze.
- bins_array: The range that defines the bins or intervals for grouping the data.

**How It Works**
    
- The function counts how many values from `data_array` fall into each bin specified by `bins_array`.
- The result is an array of frequencies:
    - Each element corresponds to the count of values in `data_array` that are less than or equal to the respective bin in `bins_array`.
    - The final element includes all values greater than the largest bin.


**Important Notes**
- The result is an array, so it must be entered as an array formula in older Excel versions (Ctrl + Shift + Enter).
- In modern Excel (Office 365 and Excel 2021), it automatically spills into adjacent cells.
- Non-numeric values in data_array are ignored.
- The bins must be in ascending order for accurate results.


```python
Scores
45
60
85
70
90
55

Intervals
50
70
90

=FREQUENCY(A1:A6, B1:B3)

# Output:
Bin	Frequency
≤ 50	2
51–70	2
71–90	2
> 90	0

# Interpreting Results
# 2 values (45, 55) are ≤ 50.
# 2 values (60, 70) fall between 51 and 70.
# 2 values (85, 90) fall between 71 and 90.
# No values are greater than 90.
```

**Applications**

- Group sales data into ranges (e.g., low, medium, high revenue).
- Analyze age groups or income brackets.
- Categorize student scores into performance tiers.


### PERCENTILE.EXC 

The PERCENTILE.EXC function calculates the value below which a given percentage (excluding the endpoints) of data in a range falls. It is commonly used in statistical analysis to determine thresholds like the 25th percentile (lower quartile), 50th percentile (median), or 90th percentile.

**Syntax:** `=PERCENTILE.EXC(array, k)`

- array: The range or array of numeric values.
- k: The percentile to calculate, expressed as a decimal between 0 and 1 (exclusive).

**Important Notes**

1. Exclusive Percentile:
    - Unlike PERCENTILE.INC, which includes 0 and 1, PERCENTILE.EXC excludes these values.
    - k must be strictly between 0 and 1 (e.g., 0.25, 0.50). Using 0, 1, or values outside this range will result in an error.

2. Interpolation:
    - If k×(n+1) (where n is the number of data points) doesn't result in an integer, Excel interpolates between the two closest ranks.

3. Sorting:
    - Excel automatically sorts the data internally, so you don't need to sort the array manually.
